```python
Випадкова величина Х має рівномірний закон розподілу, тобто ймовірності її можливих значень на інтервалі а-b однакові
За умовами завдання розподіл імовірностей Р(Х>=3)=0.4

1) Функція розподілу імовірностей:
    F(X) =  0, x <= a
            (x - a)/(b - a), a < x <= b
            1, x > b
    
    За умовами Р(Х>=3)=0.4, тобто Р(3<=Х<b)=0.4 i Р(a<Х<3)=0.6    
    
    На шкале значень Х є такі точки:  а, М(Х)=2, Х=3, b
    Імовірності Х:
        від а до Х=3: Р=0.6   
        від Х=3 до b: P=0.4   
        Тобто від а до 3 розташовано 3 умовних частини інтервалу а-b (3z)
            від 3 до b розташовано 2 умовних частини інтервалу а-b (2z)
    Є система рівнянь: 
            3-a = 3z
            b-3 =2z
        
2)  Математичне сподівання:
    М(Х) = (a + b)/2
    За умовами М(Х) = 2, тобто a + b = 4
    Відповідно b = 4 - a
    
3) Система рівнянь:
        3 - a = 3z
        (4 - a) - 3 = 2z
    Рішення:
        z = 2
        a = -3
        b = 7

4) Щільність розподілу імовірності:
    f(x) =  0, x <= a
            1/(b - a), a < x <= b
            0, x > b

Відповідь:
    f(x) =  0, x <= -3
            0.1, -3 < x <= 7
            0, x > 7    
```
